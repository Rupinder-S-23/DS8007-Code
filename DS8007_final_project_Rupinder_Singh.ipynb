```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

from matplotlib.cm import ScalarMappable
from matplotlib.lines import Line2D
from mpl_toolkits.axes_grid1.inset_locator import inset_axes
from textwrap import wrap
data = pd.read_csv("tor_players_stats.csv")
df= pd.DataFrame(data)

df_sorted=df.sort_values("PTS/G",ascending=False)
ANGLES = np.linspace(0.05, 2 * np.pi - 0.05, len(df_sorted), endpoint=False)
PTS = df_sorted["PTS/G"].values
NAMES = df_sorted["name"].values
MP = df_sorted["MP"].values
COLORS = ["#6C5B7B","#C06C84","#F67280","#F8B195"]
GREY12 = "#1f1f1f"
cmap = mpl.colors.LinearSegmentedColormap.from_list("my color", COLORS, N=256)
norm = mpl.colors.Normalize(vmin=MP.min(), vmax=MP.max())
COLORS = cmap(norm(MP))


fig, ax = plt.subplots(figsize=(12, 16), subplot_kw={"projection": "polar"})

fig.patch.set_facecolor("white")
ax.set_facecolor("white")

ax.set_theta_offset(1.2 * np.pi / 2)
ax.set_ylim(0, 25)

ax.bar(ANGLES, PTS, color=COLORS, alpha=0.9, width=0.36, zorder=10)

ax.vlines(ANGLES, 0, 30, color=GREY12, ls=(0, (4, 4)), zorder=11)

NAMES = ["\n".join(wrap(r, 5, break_long_words=False)) for r in NAMES]
NAMES

ax.set_xticks(ANGLES)
ax.set_xticklabels(NAMES, size=12)

ax.xaxis.grid(False)
ax.set_yticklabels([])
ax.set_yticks([0, 5, 10, 15, 20, 25])

ax.spines["start"].set_color("none")
ax.spines["polar"].set_color("none")

XTICKS = ax.xaxis.get_major_ticks()
for tick in XTICKS:
    tick.set_pad(10)


ax.text(-0.2 * np.pi / 2, 5 , "5", ha="center", size=12)
ax.text(-0.2 * np.pi / 2, 10, "10", ha="center", size=12)
ax.text(-0.2 * np.pi / 2, 15, "15", ha="center", size=12)
ax.text(-0.2 * np.pi / 2, 20, "20", ha="center", size=12)
ax.text(-0.2 * np.pi / 2, 25, "25", ha="center", size=12)


fig.subplots_adjust(bottom=0.11)

cbaxes = inset_axes(
    ax, 
    width="100%", 
    height="100%", 
    loc="center",
    bbox_to_anchor=(0.325, 0.1, 0.35, 0.01),
    bbox_transform=fig.transFigure 
) 

bounds = [0, 5, 10, 15, 20,25,30,35,40]
norm = mpl.colors.BoundaryNorm(bounds, cmap.N)

cb = fig.colorbar(
    ScalarMappable(norm=norm, cmap=cmap), 
    cax=cbaxes, 
    orientation = "horizontal",
    ticks=[5, 10, 15, 20,25,30,35]
)

cb.outline.set_visible(False)

cb.ax.xaxis.set_tick_params(size=0)

cb.set_label("Minutes Played Per Game", size=12, labelpad=-40)

fig.subplots_adjust(top=0.8)

title = "\nRaptors 2022 Roster Players Points Per Game"

fig.text(0.1, 0.79, title, fontsize=20, weight="bold", ha="left", va="baseline")


```
